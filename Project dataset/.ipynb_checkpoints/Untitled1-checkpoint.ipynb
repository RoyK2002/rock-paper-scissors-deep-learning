```python
import torch
import torch.nn as nn
import torch.optim as optim
import torchvision
import torchvision.transforms as transforms
import matplotlib.pyplot as plt

# Device configuration
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')

# Hyperparameters
latent_dim = 100
batch_size = 128
image_size = 32
channels_img = 3    # RGB
num_epochs = 10
lr = 0.0002
beta1 = 0.5

# Transformations
transform = transforms.Compose([
    transforms.Resize(image_size),
    transforms.ToTensor(),
    transforms.Normalize((0.5, 0.5, 0.5), (0.5, 0.5, 0.5))
])

# CIFAR10 Dataset
dataset = torchvision.datasets.CIFAR10(root='./data', download=True, transform=transform)
dataloader = torch.utils.data.DataLoader(dataset, batch_size=batch_size, shuffle=True)

# Generator Model
class Generator(nn.Module):
    def __init__(self):
        super(Generator, self).__init__()
        self.main = nn.Sequential(
            nn.ConvTranspose2d(latent_dim, 256, 4, 1, 0, bias=False),
            nn.BatchNorm2d(256),
            nn.ReLU(True),
            nn.ConvTranspose2d(256, 128, 4, 2, 1, bias=False),
            nn.BatchNorm2d(128),
            nn.ReLU(True),
            nn.ConvTranspose2d(128, 64, 4, 2, 1, bias=False),
            nn.BatchNorm2d(64),
            nn.ReLU(True),
            nn.ConvTranspose2d(64, channels_img, 4, 2, 1, bias=False),
            nn.Tanh()  # Output: (3, 32, 32)
        )

    def forward(self, x):
        return self.main(x)

# Discriminator Model
class Discriminator(nn.Module):
    def __init__(self):
        super(Discriminator, self).__init__()
        self.main = nn.Sequential(
            nn.Conv2d(channels_img, 64, 4, 2, 1, bias=False),
            nn.LeakyReLU(0.2, inplace=True),
            nn.Conv2d(64, 128, 4, 2, 1, bias=False),
            nn.BatchNorm2d(128),
            nn.LeakyReLU(0.2, inplace=True),
            nn.Conv2d(128, 256, 4, 2, 1, bias=False),
            nn.BatchNorm2d(256),
            nn.LeakyReLU(0.2, inplace=True),
            nn.Flatten(),
            nn.Linear(256*4*4, 1),
            nn.Sigmoid()
        )

    def forward(self, x):
        return self.main(x)

# Initialize models
G = Generator().to(device)
D = Discriminator().to(device)

# Loss and optimizers
criterion = nn.BCELoss()
optimizerD = optim.Adam(D.parameters(), lr=lr, betas=(beta1, 0.999))
optimizerG = optim.Adam(G.parameters(), lr=lr, betas=(beta1, 0.999))

# Training Loop
for epoch in range(num_epochs):
    for i, (real_images, _) in enumerate(dataloader):
        real_images = real_images.to(device)
        b_size = real_images.size(0)
        real_labels = torch.ones(b_size, 1, device=device)
        fake_labels = torch.zeros(b_size, 1, device=device)

        # Train Discriminator: maximize log(D(x)) + log(1 - D(G(z)))
        outputs = D(real_images)
        d_loss_real = criterion(outputs, real_labels)
        z = torch.randn(b_size, latent_dim, 1, 1, device=device)
        fake_images = G(z)
        outputs = D(fake_images.detach())
        d_loss_fake = criterion(outputs, fake_labels)
        d_loss = d_loss_real + d_loss_fake
        optimizerD.zero_grad()
        d_loss.backward()
        optimizerD.step()

        # Train Generator: maximize log(D(G(z)))
        z = torch.randn(b_size, latent_dim, 1, 1, device=device)
        fake_images = G(z)
        outputs = D(fake_images)
        g_loss = criterion(outputs, real_labels)
        optimizerG.zero_grad()
        g_loss.backward()
        optimizerG.step()

        if (i+1) % 100 == 0:
            print(f"Epoch [{epoch+1}/{num_epochs}], Step [{i+1}/{len(dataloader)}], D_loss: {d_loss.item():.4f}, G_loss: {g_loss.item():.4f}")

    # Show generated images
    if (epoch+1) % 2 == 0:
        with torch.no_grad():
            fake = G(torch.randn(16, latent_dim, 1, 1, device=device)).detach().cpu()
        grid = torchvision.utils.make_grid(fake, nrow=4, normalize=True)
        plt.figure(figsize=(6,6))
        plt.axis("off")
        plt.imshow(grid.permute(1,2,0).numpy())
        plt.show()
```

1. What did you change in the DCGAN code to accept RGB images?

I changed the input and output channels in both the Generator and Discriminator models from 1 (for grayscale) to 3 so they could process and generate RGB images. This allows the GAN to handle color images instead of just black and white.

2. Did you use the weights_init function?

No, I did not use the weights_init function anywhere in my code, as the assignment instructed me to avoid it. I relied on the default weight initialization provided by PyTorch for all layers.

3. How did you fix the normalization transformation to work with RGB images?
I used a normalization transformation that rescales all three RGB channels to have a mean of 0.5 and a standard deviation of 0.5. This adjustment ensures that the pixel values for each color channel are normalized correctly and fall within the expected range of [-1, 1].

4. Did you copy the comments around the code?

No, I did not copy any of the comments from the lecture code. Instead, I wrote my own explanations in my own words wherever necessary to describe what the code is doing.

5. How are the labels different during training the Discriminator and the Generator?

When training the Discriminator, real images from the dataset are labeled as '1' (real), and images created by the Generator are labeled as '0' (fake). However, when training the Generator, the fake images it produces are labeled as '1' (real), so the Generator tries to fool the Discriminator into thinking its generated images are real.
